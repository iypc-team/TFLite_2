```python
def augmentImages(raw_images, files, multi_factor=10):
    ''' '''
    gen=ImageDataGenerator()
    for index, image in enumerate(raw_images):
        for multi in range(multi_factor):
            print(f'files[index]: {C.BIPurple}{files[index]}{C.ColorOff}')
            theta_tfx = np.random.choice(range(270))
            print(f'theta_tfx: {theta_tfx}')
            splitExtension = files[index].split('.') # [1]
            print(splitExtension)
            fileName = split(splitExtension[0])
            print(f'fileName {fileName} {len(fileName)}')
            img_fname = contentPath + '/' + str(fileName)  + '_augmented.png'
            print(img_fname)
            
            transformed_raw_image = gen.apply_transform(
                {img, theta_tfx})
            new_image = Image.fromarray(transformed_raw_image,'RGB')
            new_image = new_image.resize((224, 224),Image.ANTIALIAS)
            new_image.save(img, img_fname)
            transformed_raw_image=None
            new_image=None
            
raw_image_dir=contentPath
raw_image_files=filesList

img_list=[]
for file in raw_image_files:
    imgArray=np.array(Image.open(file))
    img_list.append(imgArray)
    cvu.plotShowSingleImage
print(len(img_list))
augmentImages(raw_images=img_list, files=raw_image_files)
```
